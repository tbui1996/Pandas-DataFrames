```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime
import io
import requests
import numpy as np
```

```python
%matplotlib inline
r = io.StringIO(requests.get('https://db.cs.pitt.edu/courses/cs1656/data/top12cities.csv', verify=False).content.decode('utf-8'))
df = pd.read_csv(r,\
                       sep=',', engine='python')
```

```python
fig = plt.scatter(df['2014 land area'], df['2014 estimate'])

#formatting graph
plt.xlabel('land area')
plt.ylabel('2014 Estimate')
plt.title('land area and 2014 estimate')

plt.savefig("scatter_plot.png")
plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
plt.bar(df['City'], df['2014 estimate'], align='center')

plt.xlabel('City')
plt.ylabel('Population')
plt.title('City population estimate in 2014')

plt.savefig("recitation2task2.png")
plt.show()
```
